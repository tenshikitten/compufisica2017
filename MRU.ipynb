posición[i+1] = posición[i] + velocidad*Δt

```julia
Δt = 0.125
tiempo = 0:Δt:100
```

```julia
velocidad = 96
```

```julia
posición = zeros(length(tiempo));
```

```julia
for i ∈ 1:length(tiempo)
    if i != 1
        posición[i] = posición[i-1]+velocidad*Δt
    end
end
```

```julia
tiempo
```

```julia
posición
```

```julia
Tiempo = collect(tiempo)
```

```julia
plot(x=Tiempo, y=posición, Geom.line,
    Guide.xlabel("Tiempo (s)"), Guide.ylabel("Posición"), 
    Guide.title("MRU"))
```

```julia
plot(x=Tiempo, y=velocidad_arreglo, Geom.line,
    Guide.xlabel("Tiempo (s)"), Guide.ylabel("Velocidad"), 
    Guide.title("MRU"))
```

```julia
velocidad_arreglo = zeros(length(tiempo))
velocidad_arreglo = velocidad_arreglo .+ velocidad
```
